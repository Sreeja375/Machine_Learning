Datapreprocessing-Handling NA values

```python
#calculating median
median_bedrooms = df['bedrooms'].median()
print(median_bedrooms)
```

```python
#we did floor so we get integers
```

```python
df['bedrooms'] = df['bedrooms'].fillna(median_bedrooms)
df['bedrooms'] = df['bedrooms'].apply(np.floor)
df['bedrooms'] = df['bedrooms'].astype(int)
display(df)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['Area','bedrooms','age of house']],df.price)
```

```python
#predicting
reg.predict([[3000,3,40]])
```

```python
reg.coef_ #m1,m2,m3
```

```python
reg.intercept_
```

```python
#calcualted
area,age,bedrooms=3000,3,40
m1,m2,m3=reg.coef_
b=reg.intercept_
y=m1*area+m2*age+m3*bedrooms+b
y
```

implementing gradient descent


```python
def gradient_descent(x,y):
  m_curr=b_curr=0
  iterations=10
  n=len(x)
  learning_rate=0.01
  for i in range(iterations):
    y_predicted=m_curr*x+b_curr
    cost=(1/n)+sum([val**2 for val in(y-y_predicted)])#to handle engative values
    m_derivative=-(2/n)*sum(x*y-y_predicted)
    b_derivative=-(2/n)*sum(y-y_predicted)
    m_curr=m_curr-learning_rate*m_derivative
    b_curr=b_curr-learning_rate*b_derivative
    print("m {}, b{},cost{}, iteration{}".format(m_curr,b_curr,i,cost))
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
gradient_descent(x,y)

#According to our need we change iterations and learning rate
```

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Book 22(Sheet1).csv')
df
```

```python
dummies=pd.get_dummies(df.town)#to return dummy values
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
#you have to remove one of the dummy value
final=merged.drop(['town','b'], axis='columns')# b town dropped
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#x is to get all independent varaibles which are needed for trainning
#so to get all independent variables we have to drop dependent variable
X=final.drop('price',axis='columns')
X
```

```python
Y=final.price
Y
```

```python
# Drop columns where more than 50% values are NaN
X = X.dropna(thresh=len(X) // 2, axis=1)
#if there are NAN values we are unable to train model
```

```python
model.fit(X,Y)#training model
```

```python
import numpy as np
model.predict(np.array([44444,0,1]).reshape(1, -1))#predicting values
```
