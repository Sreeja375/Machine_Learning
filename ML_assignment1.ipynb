Program to perform various visualization functions

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("/content/diabetes.csv")


print(df.head())

```

```python
# 1. Histogram of Glucose levels
plt.hist(df['Glucose'], bins=20, color='skyblue', edgecolor='black')
plt.title("Glucose Distribution")
plt.xlabel("Glucose")
plt.ylabel("Count")
plt.show()
```

```python
# 2. Boxplot for BMI
sns.boxplot(x=df['BMI'])
plt.title("BMI Boxplot")
plt.show()
```

```python
# 3. Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python

# 4. Pairplot (sample of few columns for clarity)
sns.pairplot(df[['Glucose','BMI','Age','Outcome']], hue="Outcome")
plt.show()
```

```python
# Scatter Plot: Glucose vs BMI (colored by Outcome)
plt.figure(figsize=(8,6))
sns.scatterplot(x="Glucose", y="BMI", hue="Outcome", data=df, palette="Set1")
plt.title("Scatter Plot: Glucose vs BMI")
plt.show()
```

```python
# Violin Plot: Distribution of BMI grouped by Outcome
plt.figure(figsize=(8,6))
sns.violinplot(x="Outcome", y="BMI", data=df, palette="pastel", inner="quartile")
plt.title("Violin Plot: BMI by Outcome")
plt.show()
```

Program to calculate Precision and FPR

```python
from sklearn.metrics import precision_score, confusion_matrix

x = ["fish","bird","fish","fish","bird","fish","fish","bird","bird","fish"]
y = ["bird","fish","bird","bird","fish","fish","bird","bird","fish","fish"]

# Convert labels to binary (1=fish, 0=bird)
x_binary = [1 if i=="fish" else 0 for i in x]
y_binary = [1 if i=="fish" else 0 for i in y]

# Precision
precision = precision_score(y_binary, x_binary)

# Confusion matrix
tn, fp, fn, tp = confusion_matrix(y_binary, x_binary).ravel()
fpr = fp / (fp + tn)

print("Precision:", precision)
print("False Positive Rate:", fpr)
```

Program to find m and c values for iris.csv (Linear Regression)

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
df = pd.read_csv("/content/Iris.csv")
df.head()
X = df[["SepalLengthCm"]]   # independent variable
y = df["PetalLengthCm"]     # dependent variable
model = LinearRegression()
model.fit(X, y)

m = model.coef_[0]
c = model.intercept_

print("Slope (m):", m)
print("Intercept (c):", c)
```

Program to implement Multiple Linear Regression (MLR)

```python
X = df[["SepalLengthCm","SepalWidthCm"]]   # independent variable
y = df["PetalLengthCm"]     # dependent variable
model = LinearRegression()
model.fit(X, y)

m = model.coef_[0]
c = model.intercept_

print("Slope (m):", m)
print("Intercept (c):", c)
```
