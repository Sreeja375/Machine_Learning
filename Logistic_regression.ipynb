```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/insurance_data.csv')
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(df.age,df.bought_insurance,marker='+',color='red')
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df[['age']],df.bought_insurance,test_size=0.1,random_state=10)
```

```python

X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
model.predict(X_test)
```

19	18
5 	56

this is x_test from that the predicted model is  0,1,1 which means only 1st data ie age of 18 cannot buy insurance


```python
model.score(X_test,Y_test) #accuracy
```

```python
model.predict_proba(X_test)
```
