```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
#digits=load(digits)
```

```python
digits=load_digits()
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_test,Y_test)
```

for double digit the model predicting only first digit ..so to rectify this we use confusion matrix


```python
# Predict the output
y_predicted = model.predict(X_test)

# Import confusion_matrix and compute it
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_predicted)
cm
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sn
import matplotlib.pyplot as plt

# Step 1: Predict using your logistic regression model
y_predicted = model.predict(X_test)

# Step 2: Create confusion matrix
cm = confusion_matrix(Y_test, y_predicted)

# Step 3: Plot confusion matrix using seaborn
plt.figure(figsize=(10, 7))
sn.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.title('Confusion Matrix')
plt.show()
```
